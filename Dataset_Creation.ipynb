```python


import yfinance as yf
import pandas as pd
import io
import ipywidgets as widgets
from IPython.display import display, FileLink

# Function to fetch stock data
def get_stock_data(stock_symbol):
    # Get the data for the past year
    stock_data = yf.download(stock_symbol, period="2y", interval="1d")

    # Filter the required columns
    stock_data = stock_data[['Open', 'High', 'Low', 'Close']]

    return stock_data

# Function to save data as Excel
def save_to_excel(df, filename="stock_data.xlsx"):
    # Convert DataFrame to Excel
    output = io.BytesIO()
    df.to_excel(output, index=True, engine='openpyxl')
    output.seek(0)
    with open(filename, 'wb') as f:
        f.write(output.read())
    return filename

# Create input field and button in Colab
stock_input = widgets.Text(
    description="Stock Symbol:",
    value="AAPL"
)
download_button = widgets.Button(description="Download Excel")

# Define the action when the button is clicked
def on_button_click(b):
    stock_symbol = stock_input.value
    # Fetch stock data
    stock_data = get_stock_data(stock_symbol)

    # Display the head of the dataset
    display(stock_data.head())

    # Save the data to Excel and provide a link to download
    filename = save_to_excel(stock_data, f"{stock_symbol}_stock_data.xlsx")

    # Provide download link
    display(FileLink(filename))

# Link the button click event
download_button.on_click(on_button_click)

# Display the input field and button
display(stock_input, download_button)
```
