```python
import requests

url = "http://127.0.0.1:5000/fetch-indicators"
payload = {
    "symbol": "RELIANCE",
    "start_date": "01-01-2024",
    "end_date": "01-05-2024",
    "indicators": ["SMA", "RSI"]
}

response = requests.post(url, json=payload)
print("Status Code:", response.status_code)
print("Response:", response.json())
```
