```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)
y5 = np.cosh(x)
y6 = np.tanh(x)

# Create a figure with a 3x2 grid of subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(3, 2, 1)
plt.plot(x, y1, color='b')
plt.title('Sine Function')

# Second subplot
plt.subplot(3, 2, 2)
plt.plot(x, y2, color='g')
plt.title('Cosine Function')

# Third subplot
plt.subplot(3, 2, 3)
plt.plot(x, y3, color='r')
plt.ylim(-5, 5)  # Limit y-axis to avoid extreme values
plt.title('Tangent Function')

# Fourth subplot
plt.subplot(3, 2, 4)
plt.plot(x, y4, color='purple')
plt.title('Hyperbolic Sine')

# Fifth subplot
plt.subplot(3, 2, 5)
plt.plot(x, y5, color='orange')
plt.title('Hyperbolic Cosine')

# Sixth subplot
plt.subplot(3, 2, 6)
plt.plot(x, y6, color='brown')
plt.title('Hyperbolic Tangent')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the figure
plt.show()
```
