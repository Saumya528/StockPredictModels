```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 1.0
gamma = 0.5
eta = 2.0
theta = 0.1
beta = 1.0
sigma = 0.3

# Define P(t) as a function
def P(t):
    return np.sin(t)  # Example function

# Time range
t = np.linspace(0, 10, 500)
P_values = P(t)

# Compute derivatives and changes
dP_dt = np.gradient(P_values, t)  # Numerical derivative
Delta_P = np.abs(np.diff(P_values, prepend=P_values[0]))

# Piecewise max term
max_term = np.maximum(0, Delta_P - theta)

# Compute V(t)
V = alpha * np.abs(dP_dt) + gamma * sigma * P_values + eta * max_term + beta

# Plot V(t) as a function of P(t)
plt.figure(figsize=(10, 6))
plt.plot(P_values, V, label="V(P(t))")
plt.xlabel("P(t)")
plt.ylabel("V(t)")
plt.title("Graph of V(t) as a Function of P(t)")
plt.legend()
plt.grid()
plt.show()
```
