```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical data for Reliance
reliance_data = yf.download("RELIANCE.NS", start="2023-01-01", end="2024-11-11")

# Calculate moving averages
reliance_data['10-Day MA'] = reliance_data['Close'].rolling(window=10).mean()
reliance_data['20-Day MA'] = reliance_data['Close'].rolling(window=20).mean()

# Plot the Close Price and Moving Averages
plt.figure(figsize=(14, 7))
plt.plot(reliance_data.index, reliance_data['Close'], label='Close Price', color='blue')
plt.plot(reliance_data.index, reliance_data['10-Day MA'], label='10-Day MA', color='red', linestyle='--')
plt.plot(reliance_data.index, reliance_data['20-Day MA'], label='20-Day MA', color='green', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.title('Reliance Industries Stock Price and Moving Averages')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd

# Fetch stock details
end_date = datetime.datetime.now().strftime("%d-%m-%Y")
start_date = (datetime.datetime.now() - datetime.timedelta(days=365)).strftime("%d-%m-%Y")
series = "EQ"
symbol="RELIANCE"
#print(data)
# Convert data to DataFrame
df= equity_history(symbol.upper(), series, start_date, end_date)

df['Date'] = df['CH_TIMESTAMP']
df['Close'] = df['CH_CLOSING_PRICE']
df['Volume'] = df['CH_TOT_TRADED_QTY']


# Calculate MA

df['21-Day MA'] = df['Close'].rolling(window=21).mean()
df['200-Day MA'] = df['Close'].rolling(window=200).mean()

reliance_ma = df[['Close','21-Day MA','200-Day MA']].dropna()

# Display the result
print(reliance_ma)

reliance_ma.to_csv('Reliance_MovingAverage.csv')

print("Data Export Complete")
```
