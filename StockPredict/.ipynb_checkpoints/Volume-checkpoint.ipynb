 On-Balance Volume (OBV)
How it works: OBV adds volume on up days and subtracts volume on down days to create a cumulative indicator. Rising OBV signals buying pressure, while falling OBV suggests selling pressure.
Why it’s useful: OBV helps confirm price trends. If OBV rises as the price rises, it’s a sign of strength; if OBV falls during a price rise, it could indicate a weakening trend, signaling a possible reversal.
